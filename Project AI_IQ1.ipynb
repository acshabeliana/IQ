1. Tentukan Library yang akan digunakan

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
```

2. Load Dataset

```python
iq_dataset = pd.read_csv('PesertaIQ.csv')
```

```python
iq_dataset.head()
```

```python
iq_2=iq_dataset.drop(columns=['Nomor','Keterangan'])
```

```python
iq_2.head()
```

```python
iq_2.shape 
```

```python
iq_2.isnull().sum()
```

```python
iq_2['Outcome'].value_counts() 
```

```python
# memisahkan data dan label
X = iq_2.drop (columns=['Outcome', 'Nilai IQ'], axis=1)
Y = iq_2[['Outcome','Nilai IQ']]
```

```python
print(X) 
```

```python
print(Y)
```

3. Standarisasi Data

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standarized_data = scaler.transform(X) 
```

```python
print(standarized_data)
```

```python
X = standarized_data
Y=iq_2[['Nilai IQ', 'Outcome']]
```

```python
print(X)
print(Y)
```

4. Memisahkan Data Training dan Data Testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

5. Membuat data latih menggunakan algoritma Random Forest

```python
#Data untuk training dan test regresi (mendadapatkan nilai IQ)
Y_train_IQ = Y_train['Nilai IQ']
Y_test_IQ = Y_test['Nilai IQ']

print(Y_train_IQ)
```

```python
#Data untuk training dan test Klasifikasi
Y_train_outcome = Y_train['Outcome']
Y_test_outcome = Y_test['Outcome']

print(Y_train_outcome)
```

```python
regressor = RandomForestRegressor(n_estimators=100, random_state=42)
regressor.fit(X_train, Y_train_IQ)
```

```python
classifier = RandomForestClassifier(n_estimators=100, random_state=42)
classifier.fit(X_train, Y_train_outcome)
```

6. Membuat model evaluasi untuk mengukur tingkat akurasi

```python
Y_trn_IQ = regressor.predict(X_train)
mse_trn = mean_squared_error(Y_train_IQ, Y_trn_IQ)
print(f"MSE data Training Nilai IQ: {mse_trn}")
```

```python
Y_trn_outcome = classifier.predict(X_train)
accuracy = accuracy_score(Y_train_outcome, Y_trn_outcome)
print(f"Akurasi data training Outcome adalah {accuracy}")
```

```python
Y_pred_IQ = regressor.predict(X_test)
mse = mean_squared_error(Y_test_IQ, Y_pred_IQ)
print(f"MSE data testing Nilai IQ adalah {mse}")
```

```python
Y_pred_outcome = classifier.predict(X_test)
accuracy = accuracy_score(Y_test_outcome, Y_pred_outcome)
print(f"Akurasi data testing Outcome adalah {accuracy}")
```

7. Membuat model prediksi
